```python
import base64
import urllib
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
from PIL import Image
```

```python
a = open("test_portuguese.webp", "rb").read()
teste = base64.encodebytes(a)
```

```python
str(base64.b64decode(teste))
```

```python
def get_image(url):
    nparr = base64.b64decode(url)
    nparr = np.fromstring(nparr,np.uint8)
    image = cv2.imdecode(nparr, 1)
 
    return image
```

```python
imagem = get_image(teste)
```

```python
cv2_imshow(imagem)
```
